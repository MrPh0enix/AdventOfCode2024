```python
with open('input.txt', 'r') as file:
    inDoc = file.read()
```

```python
def parse_to_space(input):
    id = 0
    outStr = []
    for idx, val in enumerate(input):
        if idx % 2 == 0:
            for i in range(0, int(val)):
                outStr.append(str(id))
            id += 1
        else:
            for i in range(0, int(val)):
                outStr.append('.')
    
    return outStr

dotDoc = parse_to_space(inDoc)
print(dotDoc)
```

```python
def compact(input):
    backIdx = len(input) - 1
    output = input
    idx = 0
    # for idx, val in enumerate(output):
    while idx <= len(output)-1:
        val = output[idx]
        
        if val not in '.':
            idx += 1
            continue
        else:
            
            # find last digit
            while True:
                if output[backIdx] != '.':
                    break
                else:
                    backIdx -= 1
                
            output = output[:idx] + [output[backIdx]] + output[idx+1:backIdx]
            backIdx -= 1
        
        idx += 1
    
    return output

compactedStr = compact(dotDoc)
print(compactedStr)
            
        
```

```python
checkSum = 0



for idx, val in enumerate(compactedStr):
    
    checkSum = checkSum + (idx*int(val))

print(checkSum)
```
