```python
with open('input.txt', 'r') as file:
    inDoc = file.read()
```

```python
def parse_to_space(input):
    id = 0
    outStr = []
    for idx, val in enumerate(input):
        if idx % 2 == 0:
            # for i in range(0, int(val)):
            outStr.append([str(id)] * int(val))
            id += 1
        else:
            # for i in range(0, int(val)):
            outStr.append(['.']*int(val))
    
    return outStr

dotDoc = parse_to_space(inDoc)

dotDoc = [item for item in dotDoc if item != []]

print(dotDoc)
```

```python
import copy

def compact(input):
    backIdx = len(input) - 1
    output = copy.deepcopy(input)
    idx = 0
    
    while idx <=  len(output)-1:
        
        val = output[idx]
        
        if '.' in val:
            
            backIdx = len(output) - 1
            
            while backIdx > idx:
               
                backVal = output[backIdx]
                
                if '.' not in backVal and len(backVal) <= len(val):
                    if len(backVal) == len(val):
                        output[idx], output[backIdx] = output[backIdx], output[idx]
                        break
                    elif len(backVal) < len(val):
                        output[idx], output[backIdx] = output[backIdx], ['.']*len(backVal)
                        output.insert(idx+1, ['.']*(len(val)-len(backVal)))
                        break
                
                else:
                    backIdx -= 1
            
            idx += 1
                
        else:
            idx += 1
    
    return output

compactedStr = compact(dotDoc)
print(compactedStr)
            
        
```

```python
checkSum = 0
count = 0

for idx, val in enumerate(compactedStr):
    for idx2, val2 in enumerate(val):
        
        if '.' == val2:
            checkSum = checkSum + 0
            count += 1
        else:
            checkSum = checkSum + (count*int(val2))
            count += 1

print(checkSum)
```
