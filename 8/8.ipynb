```python
with open('input.txt', 'r') as file:
    inDoc = file.read()
    
Lis = inDoc.split('\n')
Lis = [list(x) for x in Lis]
```

```python
outSet = set()

def calcAntinodes(idx1, idx2, idx_1, idx_2):
    y_diff = idx_1 - idx1
    x_diff = idx_2 - idx2
    
    return (idx1 - y_diff, idx2 - x_diff), (idx_1 + y_diff, idx_2 + x_diff)

    

for idx1, i in enumerate(Lis):
    for idx2, j in enumerate(i):
        if j.isdigit() or j.isalpha():
            currC = j
            search = True
            idx_1 = len(Lis)-1
            idx_2 = len(Lis[0])-1
            
            while search:
                
                if idx_1 == idx1 and idx_2 == idx2:
                    break
                
                if currC == Lis[idx_1][idx_2]:
                    
                    pos1, pos2 = calcAntinodes(idx1, idx2, idx_1, idx_2)
                    
                    if -1 < pos1[0] < len(Lis) and -1 < pos1[1] < len(Lis[0]): 
                        outSet.add(pos1)
                    if -1 < pos2[0] < len(Lis) and -1 < pos2[1] < len(Lis[0]): 
                        outSet.add(pos2)
                
                idx_2 -= 1
                
                if idx_2 < 0:
                    idx_1 -= 1
                    idx_2 = len(Lis[0])-1

print(len(outSet))
```
