```python
with open('input_ord.txt', 'r') as file:
    inDoc = file.read()
    
Lis = inDoc.split('\n')
ord_dic = {}
for i in Lis:
    if int(i.split('|')[0]) not in ord_dic.keys():
        ord_dic[int(i.split('|')[0])] = [int(i.split('|')[1])]
    else:
         ord_dic[int(i.split('|')[0])].append(int(i.split('|')[1]))
```

```python
with open('input_up.txt', 'r') as file:
    inDoc = file.read()
    
Lis = inDoc.split('\n')
up_list = []
for i in Lis:
    l = i.split(',')
    up_list.append([int(x) for x in l])

len(up_list)
```

```python
count = 0
mid_elems = []
incc_ordered = []
for i in up_list:
    l=0
    r=l+1
    flag = True
    
    while l < len(i) - 1:
        while r < len(i):
            if i[r] in ord_dic[i[l]]:
                r += 1
                continue
            else:
                flag = False
                break
        l += 1
        r = l+1
        
        
    if flag == True:
        count += 1
        mid_elems.append(i[len(i)//2])
    else:
        incc_ordered.append(i)

print(f'Count: {count}')
print(f'Sum: {sum(mid_elems)}')

print(incc_ordered)
        
    
        
        
```

```python
def bubble_sort(list):
    endL = len(list) - 1
    sorting = True
    while sorting:
        sorting = False
        for i in range(0, endL):
            if list[i+1] in ord_dic[list[i]]:
                pass
            else:
                sorting = True
                list[i], list[i+1] = list[i+1], list[i]
        
    return list

sum = 0
for lis in incc_ordered:
    ret = bubble_sort(lis)
    sum += ret[len(ret)//2]
    
print(sum)
```
