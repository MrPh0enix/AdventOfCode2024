```python
import copy

with open('input.txt', 'r') as file:
    inDoc = file.read()
    
Lis = inDoc.split('\n')
Lis = [list(x) for x in Lis]
```

```python


list1 = Lis.copy()
v = 55
h = 86
state = 'UP'
outOfField = False

while not outOfField:
    if state == 'UP':
        if v-1 < 0:
            list1[v][h] = 'X'
            outOfField = True
            continue
        elif list1[v-1][h] == '#':
            state = 'RIGHT'
            continue
        elif list1[v-1][h] in ['.', 'X', '^']:
            list1[v][h] = 'X'
            v -= 1
            continue
    elif state == 'RIGHT':
        if h+1 >= len(list1[0]) :
            list1[v][h] = 'X'
            outOfField = True
            continue
        elif list1[v][h+1] == '#':
            state = 'DOWN'
            continue
        elif list1[v][h+1] in ['.', 'X', '^']:
            list1[v][h] = 'X'
            h += 1
            continue
    elif state == 'DOWN':
        if v+1 >= len(list1) :
            list1[v][h] = 'X'
            outOfField = True
            continue
        elif list1[v+1][h] == '#':
            state = 'LEFT'
            continue
        elif list1[v+1][h] in ['.', 'X', '^']:
            list1[v][h] = 'X'
            v += 1
            continue
    elif state == 'LEFT':
        if h-1 < 0:
            list1[v][h] = 'X'
            outOfField = True
            continue
        elif list1[v][h-1] == '#':
            state = 'UP'
            continue
        elif list1[v][h-1] in ['.', 'X', '^']:
            list1[v][h] = 'X'
            h -= 1
            continue
        

count = 0
for idx1, i in enumerate(list1):
    for idx2, j in enumerate(i):
        if j in ['X', '^']:
            count += 1
            
print(count)          
            

```
