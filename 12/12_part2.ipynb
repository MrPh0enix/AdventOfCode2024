```python
with open('input.txt', 'r') as file:
    Doc = file.read()
    
Lis = Doc.split('\n')
finLis = [list(i) for i in Lis]
finLis
```

```python
import copy
outLis = []

def convert_to_ids(input):
    global outLis
    id = 0
    outLis = copy.deepcopy(input)
    for idx1, val1 in enumerate(outLis):
        for idx2, val2 in enumerate(val1):
            
            if not isinstance(val2, tuple):
                val = outLis[idx1][idx2]
                recurStar(idx2, idx1, outLis, val, id, input)
                id += 1
                
    return outLis


def recurStar(x,y, outLis, val, id, input):
    
    sidesLis = checkSides(x, y, val, input)
    outLis[y][x] = (str(id), sidesLis)
    
    #UP
    if y > 0 and not isinstance(outLis[y-1][x], tuple) and outLis[y-1][x] == val:
        recurStar(x, y-1, outLis, val, id, input)
    #DOWN
    if y < len(outLis)-1 and not isinstance(outLis[y+1][x], tuple) and outLis[y+1][x] == val:
        recurStar(x, y+1, outLis, val, id, input)
    #LEFT
    if x > 0 and not isinstance(outLis[y][x-1], tuple) and outLis[y][x-1] == val:
        recurStar(x-1, y, outLis, val, id, input)
    #RIGHT
    if x < len(outLis[0])-1 and not isinstance(outLis[y][x+1], tuple) and outLis[y][x+1] == val:
        recurStar(x+1, y, outLis, val, id, input)
        
    return


def checkSides(x, y, val, input):
    sidesLis = []
    
    if y == 0 or (y > 0 and input[y-1][x] != val):
        sidesLis.append('Up')
    if y == len(input)-1 or (y < len(input)-1 and input[y+1][x] != val):
        sidesLis.append('Dw')
    if x == 0 or (x > 0 and input[y][x-1] != val):
        sidesLis.append('Lt')
    if x == len(input[0])-1 or (x < len(input[0])-1 and input[y][x+1] != val):
        sidesLis.append('Rt')
    
    return sidesLis


numLis = convert_to_ids(finLis)
numLis
```

```python
doneLis = []

def calcPrice(input):
    global doneLis
    numsDone = []
    score = 0
    doneLis = copy.deepcopy(input)
    for idx1, val1 in enumerate(input):
        for idx2, val2 in enumerate(val1):
            
            if  val2[0] not in numsDone:
                numsDone.append(val2[0])
                sidesLis = []
                area, sidesLis = calcArea(idx1, idx2, doneLis, val2[0], 0, input, [], sidesLis)
                sides = calcSides(sidesLis)
                score += (area*sides)
    
    return score

            
def calcArea(y, x, doneLis, val, area, input, prevSides, sidesLis):
    
    currSides = doneLis[y][x][1]
    
    for i in currSides:
        sidesLis.append([i , (x, y)])
    
    area += 1
    
    doneLis[y][x] = '*'
    
    
    #UP
    if y > 0 and isinstance(doneLis[y-1][x], tuple) and doneLis[y-1][x][0] == val:
        
        area, sidesLis = calcArea(y-1, x, doneLis, val, area, input, currSides, sidesLis)
    #DOWN
    if y < len(doneLis)-1 and isinstance(doneLis[y+1][x], tuple) and doneLis[y+1][x][0] == val:
        
        area, sidesLis = calcArea(y+1, x, doneLis, val, area, input, currSides, sidesLis)
    #LEFT
    if x > 0 and isinstance(doneLis[y][x-1], tuple) and doneLis[y][x-1][0] == val:
        
        area, sidesLis = calcArea(y, x-1, doneLis, val, area, input, currSides, sidesLis)
    #RIGHT
    if x < len(doneLis[0])-1 and isinstance(doneLis[y][x+1], tuple) and doneLis[y][x+1][0] == val:
        
        area, sidesLis = calcArea(y, x+1, doneLis, val, area, input, currSides, sidesLis)    
    
    
    return area, sidesLis



def calcSides(sidesLis):
    sides = 0
    
    upLis = []
    dwLis = []
    ltLis = []
    rtLis = []
    
    for i in sidesLis:
        if i[0] == 'Up':
            upLis.append(i[1])
        if i[0] == 'Dw':
            dwLis.append(i[1])
        if i[0] == 'Lt':
            ltLis.append(i[1])
        if i[0] == 'Rt':
            rtLis.append(i[1])
    
    
    # UP List 
    idx = 0
    increment = 1
    while idx < len(upLis)-1:
        
        if upLis[idx] != '*':
            (x, y) = upLis[idx]
            upLis[idx] = '*'
            increment = 1
            decrement = 1
            while True:
                if (x+increment, y) in upLis:
                    upLis.remove((x+increment, y))
                    increment += 1
                else:
                    break
            while True:
                if (x-decrement, y) in upLis:
                    upLis.remove((x-decrement, y))
                    decrement += 1
                else:
                    break
            idx += 1
                
        else:
            idx += 1  
    
    sides += len(upLis) 
    
    
    # Down List 
    idx = 0
    increment = 1
    while idx < len(dwLis)-1:
        
        if dwLis[idx] != '*':
            (x, y) = dwLis[idx]
            dwLis[idx] = '*'
            increment = 1
            decrement = 1
            while True:
                if (x+increment, y) in dwLis:
                    dwLis.remove((x+increment, y))
                    increment += 1
                else:
                    break
            while True:
                if (x-decrement, y) in dwLis:
                    dwLis.remove((x-decrement, y))
                    decrement += 1
                else:
                    break
            idx += 1
                
        else:
            idx += 1  
    
    sides += len(dwLis)
    
    
    # Left List 
    idx = 0
    increment = 1
    while idx < len(ltLis)-1:
        
        if ltLis[idx] != '*':
            (x, y) = ltLis[idx]
            ltLis[idx] = '*'
            increment = 1
            decrement = 1
            while True:
                if (x, y+increment) in ltLis:
                    ltLis.remove((x, y+increment))
                    increment += 1
                else:
                    break
            while True:
                if (x, y-decrement) in ltLis:
                    ltLis.remove((x, y-decrement))
                    decrement += 1
                else:
                    break
            idx += 1
                
        else:
            idx += 1  
    
    sides += len(ltLis)
    
    
    # Right List 
    idx = 0
    increment = 1
    while idx < len(rtLis)-1:
        
        if rtLis[idx] != '*':
            (x, y) = rtLis[idx]
            rtLis[idx] = '*'
            increment = 1
            decrement = 1
            while True:
                if (x, y+increment) in rtLis:
                    rtLis.remove((x, y+increment))
                    increment += 1
                else:
                    break
            while True:
                if (x, y-decrement) in rtLis:
                    rtLis.remove((x, y-decrement))
                    decrement += 1
                else:
                    break
            idx += 1
                
        else:
            idx += 1  
    
    sides += len(rtLis)  
    
    
    return sides
    
        

    
    
    
            
            
print(calcPrice(numLis))
```
